```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
import numpy as np

# Charger mod√®le entra√Æn√©
model = YOLO('runs/detect/train/weights/best.pt')

# Image test
image_path = '/content/drive/MyDrive/projet_yolo_medical/data/images/test/65.jpg'

# Pr√©diction
results = model.predict(source=image_path, conf=0.25)
result = results[0]

# Extraire donn√©es
boxes = result.boxes
cls_ids = boxes.cls.cpu().numpy().astype(int)
conf_scores = boxes.conf.cpu().numpy()
xyxy = boxes.xyxy.cpu().numpy()

# Classes et couleurs (modifie selon ton jeu de classes)
class_names = ['microaneurysms', 'haemorrhages', 'hard_exudates', 'soft_exudates', 'optic_disc']
class_colors = {
    0: (255, 0, 0),      # rouge
    1: (0, 255, 0),      # vert
    2: (0, 0, 255),      # bleu
    3: (255, 255, 0),    # jaune
    4: (255, 0, 255),    # magenta
}

# Classes d√©tect√©es (uniques)
present_classes = np.unique(cls_ids)

# Lire image originale (RGB)
img_bgr = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Pr√©parer la figure avec autant de subplots que classes d√©tect√©es
n = len(present_classes)
fig, axs = plt.subplots(n, 1, figsize=(6*n+5, 6+5))
if n == 1:
    axs = [axs]

for i, class_id in enumerate(present_classes):
    ax = axs[i]
    img_copy = img_rgb.copy()

    # Indices des boxes pour cette classe
    idxs = np.where(cls_ids == class_id)[0]

    # Dessiner uniquement les boxes de cette classe
    for idx in idxs:
        x1, y1, x2, y2 = xyxy[idx].astype(int)
        conf = conf_scores[idx]
        color = class_colors[class_id]
        label = f"{class_names[class_id]} {conf:.2f}"

        cv2.rectangle(img_copy, (x1, y1), (x2, y2), color, 2)
        cv2.putText(img_copy, label, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.3, color, 2)

    ax.imshow(img_copy)
    ax.axis('off')
    ax.set_title(class_names[class_id])

plt.tight_layout()
plt.show()
```
